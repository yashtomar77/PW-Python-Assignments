# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

LookupError Exception is the Base class for errors raised when something can't be found. It is a subclass of the Exception class and serves as a common base class for more specific lookup-related errors in Python. Two examples of such errors are KeyError and IndexError. Let's understand with a examaple

```python
# 1. KeyError: This error occurs when a dictionary 'key' or a set element is not found.

dictionary = {'name': 'yash', 'age': 22}
try:
    value = dictionary['qualification']
    value
except KeyError as e:
    print("Error: age not found" ,e)
```

```python
# 2. IndexError: This error occurs when an 'index' is out of range for a sequence like a list or a string.

l = [10,20,30,40,50]
try:
    result = l[7]
    result
except IndexError as f:
    print('Error:{}'.format(f))
```

# Q5. Explain ImportError. What is ModuleNotFoundError?

'ImportError' is an exception that is raised when an import statement fails to import a module or when a module fails to import a specific attribute. It is a subclass of the 'Exception' class and is commonly encountered when working with Python modules and packages.

'ModuleNotFoundError', it is a subclass of ImportError that specifically indicates that a requested module could not be found or imported.

Both 'ImportError' and 'ModuleNotFoundError' serve similar purposes of signaling import-related errors. However, ModuleNotFoundError provides more clarity
